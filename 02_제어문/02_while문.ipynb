```python
# while문 기본 구조

# while 조건문:
# 	수행할문장1
# 	수행할문장2
# 	수행할문장3
#	...
```

```python
treehit = 0
while treehit < 10:
	treehit = treehit + 1
	print(f"나무를 {treehit}번 찍었습니다.")
	if treehit == 10:
		print("나무 넘어갑니다.")
```

```python
number = 0
prompt = """
1. Add
2. Del
3. List
4. Quit
"""
while number != 4:
	print(prompt)
	number = int(input("1~4까지의 숫자중 하나를 입력해주세요"))
```

```python
# while문 강제로 빠져나가기
커피 = 5
돈 = 300
while 돈:
	print("돈을 받았으니 커피를 줍니다")
	커피 -= 1 #커피 = 커피 - 1
	print(f"남은 커피는 {커피}개 입니다.")
	if 커피 == 0:
		print("커피를 다 팔았습니다. 판매를 중지합니다")
		break
```

```python
coffee = 10

while True:
    try:
        money = int(input("돈을 넣어 주세요: "))
    except:
        print("숫자만 입력하세요!")
        continue

    if money == 300:
        print("커피를 줍니다.")
        coffee -= 1

    elif money > 300:
        print(f"거스름돈 {money - 300}원을 주고, 커피를 줍니다.")
        coffee -= 1

    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")

    print(f"현재 남은 커피는 {coffee}개 입니다")

    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
# while문의 맨 처음으로 돌아가기 - continue
# 1~10중에서 홀수만 출력하기

n = 0
while n < 10: 
	n = n + 1 # n += 1
	if n % 2 == 0:
		continue
	print(n)
	
```

```python
# for문의 기본구조
# for 변수 in 리스트(또는 튜플, 문자열):
# 	수행할 문장1
# 	수행할 문장2
# 	...

num_list = ["one", "two", "three"]
for num in num_list:
	print(num)
```

```python
tuple_list = [(1,2), (3,4), (5,6)]
for a in tuple_list:
	print(a)
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
	number = number + 1
	if mark >= 60:
		print(f"{number}학생은 통과입니다.")
	else:
		print(f"{number}학생은 불합격입니다.")
```

```python
# for문과 continue문
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks :
    number += 1
    if mark < 60 :
        continue
    else :
        print(f"{number}학생은 합격합니다.")
```

```python
# for문과 함께 자주 사용하는 range함수
a = range(10)
print(a)
a = range(0, 10)
print(list(a))
```

```python
result = 0
for number in range(1, 10+1):
	result = result + number # result += number

print(result)
```

```python
marks = [90, 25, 67, 45, 80]

for number in range(len(marks)):
	if marks[number] < 60:
		continue
	print(f"{number+1} 학생은 합격입니다.")
```

```python
# for와 range를 이용한 구구단
for i in range(2, 10):
	for j in range(1, 10):
		print(i * j, end=" ")
	print("")
```

```python
# 출력결과를
# 2 x 1 = 2
# 2 x 2 = 4
# ...
# 2 x 9 = 18

# 3 x 1 = 3
# ...
# ...

# 9 x 9 = 81

for i in range(2, 10):       
    for j in range(1, 10):   
        print(f"{i} x {j} = {i * j}")
    print()

```

```python
# *
# **
# ***
# ****
# *****
for i in range(1, 6):
	print("*" * i)
print()
```

```python
#     *
#    **
#   ***
#  ****
# *****
for i in range(1, 6):
    print(" " * (5 - i) + "*" * i)
```

```python
# 리스트 컴프리헨션 사용하기
a = [1, 2, 3, 4]
result = []
for num in a:
	result.append(num * 3)

print(result)
```

```python
a = [1, 2, 3, 4]

result = [num * 3 for num in a]
print(result)
```

```python
# for문과 break문
for i in range(10):
	if i == 5:
		break
	print(i)
```

```python
# enumerate()함수 활용
fruits = ["apple", "banana", "orange"]
for i, fruit in enumerate(fruits):
	print(f"{i}번째 : {fruit}")
```

```python
# zip함수로 여러 리스트 함께 순히하기
names = ["홍길동", "유재석", "강호동"]
scores = [60, 70, 80]
for name, score in zip(names, scores):
	print(f"{name}: (score)점")
```

```python
# 3개 이상의 리스트 - zip()
names = ["홍길동", "유재석", "강호동"]
Korean = [60, 70, 80]
English = [80, 90, 70]

for name, score1, score2 in zip(names, Korean, English):
	print(f"{name}: 국어:{score1}, 영어:{score2}")
```
