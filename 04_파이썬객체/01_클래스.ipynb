```python
class Calculator:
	def __init__(self):
		self.result = 0

	def add(self, num):
		self.result = self.result + num 
		return self.result 
	
	def sub(self, num):
		self.result -= num
		return self.result
	
	def mul(self, num):
		self.result *= num
		return self.result
	
	def div(self, num):
		self.result /= num
		return self.result

cal1 = Calculator()
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

cal2 = Calculator()
print(cal2.add(5))
print(cal2.sub(10))
print(cal2.mul(10))
print(cal2.div(5))


```

```python
# 객체에 연산할 숫자 지정하기

class FourCal:
	def setdata(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
		

cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
```

```python
class FourCal:
	def setdata(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
	def sub(self):
		result = self.first - self.second
		return result
	def mul(self):
		result = self.first * self.second
		return result
	def div(self):
		result = self.first / self.second
		return result
	
print(cal3.first)
print(cal3.second)
print(cal3.add())
```
