```python
# 파이썬 함수의 구조
# def 함수이름(매개변수1, 매개변수2, ...):
# 	수행할 문장1
# 	수행할 문장2
# 	...
	
def add(a, b):
	return a + b
```

```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다

def add(a, b):
	result = a + b
	return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값이 없는 함수
# 입력값 X, 반환값 O

def say():
	return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 입력값 O, 반환값 X
def add(a, b):
	print(f"{a}, {b}의 합은 {a+b}입니다.")

sum = add(3, 4)
print(sum)
```

```python
# 입력값도 없고, 반환값도 없는 함수
# 입력값 X, 반환값 X

def say():
	print("hi")

a = say()
print(a)
```

```python
products = {
    "1": {"name": "사과", "price": 2000},
    "2": {"name": "바나나", "price": 3000},
    "3": {"name": "우유", "price": 2500},
    "4": {"name": "빵", "price": 4000},
    "5": {"name": "치즈", "price": 5000}
}

cart = []

while True:
    print("\n1.상품목록 2.장바구니담기 3.장바구니보기 4.결제 5.종료")
    menu = input("선택: ")
    
    if menu == "1":
        for key, val in products.items():
            print(f"{key}. {val['name']} : {val['price']:,}원")
    
    elif menu == "2":
        item_no = input("담을 상품 번호: ")
        if item_no in products:
            cart.append(products[item_no])
            print(f"{products[item_no]['name']}이(가) 담겼습니다.")
        else:
            print("없는 상품입니다.")
            
    elif menu == "3":
        total = sum(item['price'] for item in cart)
        print("\n[장바구니 내역]")
        for item in cart:
            print(f"- {item['name']}: {item['price']:,}원")
        print(f"총 예상 금액: {total:,}원")
        
    elif menu == "4":
        if not cart:
            print("장바구니가 비어있습니다.")
            continue
            
        total = sum(item['price'] for item in cart)
        coupon = input("10% 할인 쿠폰을 사용하시겠습니까? (y/n): ")
        
        if coupon.lower() == 'y':
            total = int(total * 0.9)
            print("쿠폰이 적용되었습니다.")
            
        if total < 30000:
            total += 3000
            print("배송비 3,000원이 추가되었습니다. (30,000원 이상 무료배송)")
            
        print(f"최종 결제 금액은 {total:,}원입니다. 감사합니다.")
        cart.clear()
        
    elif menu == "5":
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다.")
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 할까?
def add_many(*args):
	result = 0
	print(type(args))
	print(args)
	for i in args:
		result = result + i
	return result

result = add_many(1, 2, 3, 4, 5)
print(result)
print()
```

```python
def add_mul(choice, *args):
	if choice == "add":
		result = 0
		for i in args:
			result = result + i


	elif choice == "mul":
		result = 1
		for i in args:
			result = result * i

	return result
```

```python
result = add_mul("add", 1, 2, 3, 4, 5)
print(result)
```

```python
result = add_mul("mul", 1, 2, 3, 4, 5)
print(result)
```

```python
def say_nick(nick):
	if nick == "학생":
		return
	print(f"너의 별명은 {nick}이다")

say_nick("선생")
say_nick("학생")
```

```python
# 매개변수에 초깃값 미리 설정하기

def say_myself(name, age, man=True):
	print(f"너의 이름은 {name}이다")
	print(f"너의 나이은 {age}이다")
	if man:
		print("남자입니다")
	else:
		print("여자입니다")

say_myself("홍길동", 25)
say_myself("김영희", 30, False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위

a = 1
def vartest(a):
	a = a + 1

vartest(a)
print(a)
```

```python
a = 1
def vartest():
	global a
	a = a + 1

vartest()
print(a)
```

```python
# lambda 예약어
# 함수_이름 = lambda 매개변수1, 매개변수2, ... : 수행할 문장
def add (a, b):
	return a + b


add = lambda a, b: a + b
```

```python
# 함수의 독스트림(Docstring)

def add(a, b):
	"""
	두 숫자를 더하는 함수

	Parameters:
	a(int, float): 첫 번째 숫자
	b(int, float): 두 번째 숫자

	returns:
	int, float: 두 숫자의 합

	"""
	return a + b


print(add.__doc__)
```
