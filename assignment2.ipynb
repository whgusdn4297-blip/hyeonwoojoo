```python
text = """I really enjoy learning python programming because it is powerful and easy to use"""

length = len(text)
print(length)

length_no_spaces = len(text.replace(" ", ""))
print(length_no_spaces)

uppercase_text = text.upper()
print(uppercase_text)

word_count = len(text.split())
print(word_count)

contains_python = "python" in text.lower()
print(contains_python)
```

```python
items = ["a", "b", "c", "b", "d", "b", "e"]

items.remove("b")
print(items)

removed_item = items.pop()
print(f"제거된 항목: {removed_item}")
print(items)

removed_item = items.pop(0)
print(f"제거된 항목: {removed_item}")
print(items)

items.clear()
print(items)
```

```python
numbers = [64, 34, 25, 12, 22, 11, 90]
words = ["banana", "Apple", "cherry", "date"]

numbers.sort()
print(numbers)

numbers.sort(reverse=True)
print(numbers)

words.sort()
print(words)

words.reverse()
print(words)
```
